```python

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
from  scipy.optimize import minimize
from  sklearn.metrics   import r2_score
from  numpy.random      import default_rng
import statsmodels.api as sm
from pathlib import Path


#Load GDP data
BASE= Path('/Users/antoinecyrusbecharat')

df_gdp=pd.read_csv(BASE /'downloads/GDP_us_long.csv')

#filter GDP trend
cycle, trend = sm.tsa.filters.hpfilter(df_gdp['GDP'], lamb=1600)
plt.plot(cycle)
plt.show()
    
win=120

mu  = cycle.rolling(win, center=True, min_periods=1).mean()
sig = cycle.rolling(win, center=True, min_periods=1).std(ddof=0)

#standardize GDP
gdp = (cycle - mu) / sig

    
#load US unemployment/vacancy data

df_uv = pd.read_csv(BASE / "Labor_Network_INET/us_bev_curve.csv")


# Build a monthly DateTimeIndex for df_uv starting 1951-01-01

n_months = df_uv.shape[0]
uv_idx = pd.date_range(start="1951-01-01", periods=n_months, freq="MS")
df_uv.index = uv_idx

#Interpolate GDP to the same months as df_uv

gdp_sub = gdp[16:-4]
print(gdp_sub)


gdp_monthly = pd.Series(
    np.interp(
        np.linspace(0, len(gdp_sub)-1, len(df_uv)),
        np.arange(len(gdp_sub)),
        gdp_sub.values
    ),
    index=df_uv.index,
    name="GDP_monthly"
)




print(df_uv)
print(gdp_monthly)

#merge into a single data frame
df_all = pd.concat([df_uv, gdp_monthly], axis=1).dropna()


#display the time series

plt.figure(figsize=(12, 5))
plt.plot(df_all.index, df_all["unemployment rate"], label="Unemployment rate")
plt.plot(df_all.index, df_all["vacancy rate"],      label="Vacancy rate")
plt.plot(df_all.index, df_all["GDP_monthly"],       label="GDP (interpolated)")

plt.legend(loc="upper left")
plt.title("Monthly Unemployment, Vacancy Rate, and Interpolated GDP")
plt.xlabel("Date")
plt.tight_layout()
plt.show()
```

```python
# Simulate the model with sine GDP to reproduce the 2008 business cycle



df_emp = df_all.loc["2000-01":"2018-12"].copy()
u_emp = df_emp["unemployment rate"].to_numpy() / 100   #the raw data is in percentage
v_emp = df_emp["vacancy rate"].to_numpy() / 100
g_emp = df_emp["GDP_monthly"].to_numpy() / 100
dates = df_emp.index
N_emp = len(u_emp)
dt = 1.0
period = 100

# Sine synthetic GDP 
def synthetic_gdp_sine_matched(emp_g, period=100, phase=0, cycles=10):
    N = len(emp_g) * cycles
    mean = np.mean(emp_g)
    emp_dg = np.gradient(emp_g)
    ampl = np.std(emp_dg) * period / (2 * np.pi)
    t = np.arange(N)
    g_synth = mean + ampl * np.sin(2 * np.pi * t / period + phase)
    return g_synth

cycles = 10 #we simulate the model for 10 business cycles
g_synth = synthetic_gdp_sine_matched(g_emp, period=period, phase=0, cycles=cycles)


shift_months=24 #we find empirically that we need to shift gdp by 24 months to align well with the empirical data
g_used = np.concatenate((g_synth[shift_months:], np.full(shift_months, g_synth[-1])))
dgdt_synth = np.gradient(g_used, dt)

#Simulate coupled u and v using synthetic GDP 
def simulate_uv(theta, dgdt_synth):
    s, K, C_u, v0, C_v = theta
    N = len(dgdt_synth)
    u = np.empty(N); v = np.empty(N); m = np.empty(N)
    u[0], v[0] = u_emp[0], v_emp[0]
    m[0] = K * u[0] * v[0]
    for t in range(1, N):
        g_neg = -max(-dgdt_synth[t-1], 0.0)
        du = (1 - u[t-1]) * (s - C_u * g_neg) - K * u[t-1] * v[t-1]
        dv = (1 - v[t-1]) * (v0 + C_v * dgdt_synth[t-1]) - K * u[t-1] * v[t-1]
        u[t] = np.clip(u[t-1] + dt * du, 0, 1)
        v[t] = np.clip(v[t-1] + dt * dv, 0, 1)
        m[t] = K * u[t] * v[t]
    return u, v, m

# Use fixed parameters (the value for s and v0 were found by fitting the data using the coupled model) 
#we take K = C_u = C_v here 

theta_fixed = [0.0084, 0.5, 2, 0.0089, 2]



u_sim_full, v_sim_full, m_sim_full = simulate_uv(theta_fixed, dgdt_synth)

# Keep only cycles 7 to 9 for plotting, to ensure that we are not in a transient

plot_range = slice(6 * period, 9 * period)
u_sim = u_sim_full[plot_range]
v_sim = v_sim_full[plot_range]
m_sim = m_sim_full[plot_range]
g_used_plot = g_used[plot_range]
dates = pd.date_range("2000-01", periods=len(u_sim), freq="MS")

# Plot u, v, and GDP
fig, ax1 = plt.subplots(figsize=(11, 5))
ax1.plot(dates, u_sim * 100, label="Unemployment (%)", color="C0")
ax1.plot(dates, v_sim * 100, label="Vacancies (%)", color="C1")
ax1.set_ylabel("Rate (%)")

# Identify extrema within 2004-2012 range
mask_ext = (dates >= "2004-01") & (dates <= "2012-12")
idx_ext = np.where(mask_ext)[0]
u_range = u_sim[idx_ext]
v_range = v_sim[idx_ext]



ax2 = ax1.twinx()
ax2.plot(dates, g_used_plot, label="Synthetic GDP", color="C2")
ax2.set_ylabel("GDP")

lines_1, labels_1 = ax1.get_legend_handles_labels()
lines_2, labels_2 = ax2.get_legend_handles_labels()
ax1.legend(lines_1 + lines_2, labels_1 + labels_2, loc="upper left", frameon=False)
plt.title("Synthetic GDP, Unemployment, Vacancies, C_u=C_v=2, K=0.5")
plt.tight_layout()
#plt.savefig("/Users/antoinecyrusbecharat/Labor_Network_INET/synth_gdp_uv_Cu_equal_Cv_K_small.pdf",bbox_inches="tight")
plt.show()

# Plot u, v, and inverse matching 
inv_match = 1.0 / m_sim
fig, ax1 = plt.subplots(figsize=(11, 5))
ax1.plot(dates, u_sim * 100, label="Unemployment (%)", color="C0")
ax1.plot(dates, v_sim * 100, label="Vacancies (%)", color="C1")
ax1.set_ylabel("Rate (%)")


ax2 = ax1.twinx()
ax2.plot(dates, inv_match, label=r"$1/(Kuv)$", color="k", linestyle="--")
ax2.set_ylabel("Inverse Matching Rate")

lines_1, labels_1 = ax1.get_legend_handles_labels()
lines_2, labels_2 = ax2.get_legend_handles_labels()
ax1.legend(lines_1 + lines_2, labels_1 + labels_2, loc="upper left", frameon=False)
plt.title("Synthetic Inverse Matching Rate, Unemployment, Vacancies, increased Cu and lowered K")
plt.tight_layout()
#plt.savefig("/Users/antoinecyrusbecharat/Labor_Network_INET/synth_gdp_uv_matching_high_Cu_low_K.pdf",bbox_inches="tight")
plt.show()

# Beveridge curve 

x = u_sim*100
y = v_sim*100

fig, ax = plt.subplots(figsize=(6,6))
ax.plot(x, y, '-k', lw=1)
step = max(1, len(x)//10)
for i in range(0, len(x)-step, step):
    dx, dy = x[i+1]-x[i], y[i+1]-y[i]
    ax.arrow(x[i], y[i], dx, dy,
             shape='full', lw=0,
             length_includes_head=True, head_width=0.2, color='k')

ax.set_xlabel("Unemployment (%)")
ax.set_ylabel("Vacancies (%)")
ax.set_title("Simulated Beveridge Curve")
plt.tight_layout()
#plt.savefig("/Users/antoinecyrusbecharat/Labor_Network_INET/synth_gdp_bev_curve_T100_Cu10_K2_Cv10.pdf", bbox_inches="tight")
plt.show()

x=u_sim * 100
y=v_sim * 100
# compute area enclosed by the Beveridge curve
area = 0.5 * np.abs(np.sum(x[:-1]*y[1:] - x[1:]*y[:-1]))
print(f"Area enclosed by Beveridge curve: {area:.3f} (%²)")
```

```python
# Simulate the model with negatively skewed GDP to reproduce the skewness behavior observed for Delta u and Delta v


from scipy.stats import skew

df_emp = df_all.loc["2000-01":"2018-12"].copy()
u_emp = df_emp["unemployment rate"].to_numpy() / 100
v_emp = df_emp["vacancy rate"].to_numpy() / 100
g_emp = df_emp["GDP_monthly"].to_numpy() / 100

# === 1) Parameters ===
dt      = 1.0
period  = 100    # cycle length
cycles  = 10
burn_g  = 200    # burn-in for GDP (again, to prevent transients)
burn_uv = 200    # burn-in for u,v
alpha   = 0.05   # noise amplitude
df_nt   = 2      # Student-t degrees of freedom (perhaps try df=1 ?)

# Empirical moments
mean_g = np.mean(g_emp)
std_dg = np.std(np.gradient(g_emp, dt))

# AR(2) parameters for simulated GDP
r     = 0.99
phi1  = 2*r*np.cos(2*np.pi/period)
phi2  = -r**2
N     = len(g_emp)*cycles + burn_g

# Generate negatively skewed GDP series
g = np.full(N, mean_g)
rng = np.random.default_rng(42)
# student-t noise, then take negative absolute for negative skew
eps = -np.abs(rng.standard_t(df_nt, size=N)) * std_dg * alpha
for t in range(2, N):
    g[t] = (
        mean_g
        + phi1 * (g[t-1] - mean_g)
        + phi2 * (g[t-2] - mean_g)
        + eps[t]
    )

g_ar = g[burn_g:]
shift = 24
g_used = np.concatenate((g_ar[shift:], [g_ar[-1]]*shift))
dgdt = np.gradient(g_used, dt)

# Simulate the model
def simulate_uv(theta, dgdt, burn=burn_uv):
    s, K, C_u, v0, C_v = theta
    N = len(dgdt)
    u = np.zeros(N+burn)
    v = np.zeros(N+burn)
    u[:burn] = u_emp[-1]
    v[:burn] = v_emp[-1]
    for t in range(burn+1, N+burn):
        i = t - burn
        # positive response to GDP drop
        neg_dg = -max(-dgdt[i-1], 0)
        du = (1-u[t-1])*(s - C_u*neg_dg) - K*u[t-1]*v[t-1]
        dv = (1-v[t-1])*(v0 + C_v*dgdt[i-1]) - K*u[t-1]*v[t-1]
        u[t] = np.clip(u[t-1] + dt*du, 0, 1)
        v[t] = np.clip(v[t-1] + dt*dv, 0, 1)
    return u[burn:], v[burn:]

#  Run and extract last cycle 
theta   = [0.0084, 2, 2, 0.0089, 2.0]  
u_sim, v_sim = simulate_uv(theta, dgdt)

N_emp = len(g_emp)
# last cycle indices
start =  (cycles-2)*N_emp
end   = cycles * N_emp

# convert to percentages
u_win = u_sim[start:end] * 100
v_win = v_sim[start:end] * 100
g_win = g_used[start:end]

# === 5) Plot Beveridge curve for last cycle ===
plt.figure(figsize=(6,5))
plt.plot(u_win, v_win, lw=1.5)
plt.xlabel("Unemployment (%)")
plt.ylabel("Vacancy (%)")
plt.title("Beveridge Curve (with AR(2) gdp)")
plt.tight_layout()
#plt.savefig("/Users/antoinecyrusbecharat/Labor_Network_INET/synth_gdp_bev_curve_ar2.pdf",bbox_inches="tight")
plt.show()


```

```python
# Using the exact same procedure, we now simulate for several values of C_u
# Doyne suggested that C_u < 1 (which is what we find empirically)


df_emp = df_all.loc["2000-01":"2018-12"].copy()
u_emp  = df_emp["unemployment rate"].to_numpy() / 100
v_emp  = df_emp["vacancy rate"].to_numpy()      / 100
g_emp  = df_emp["GDP_monthly"].to_numpy()       / 100

# Global parameters 
dt        = 1.0
period    = 100
cycles    = 40
burn_g    = 2000        # GDP burn-in
burn_uv   = 2000        # u,v burn-in

mean_g    = g_emp.mean()
std_dg    = np.gradient(g_emp, dt).std()

# AR(2) parameters for simulated GDP 

alpha   = 0.05   # noise amplitude
df_nt   = 2      # Student-t degrees of freedom (perhaps try df=1 ?)
r         = 0.99
phi1      = 2 * r * np.cos(2 * np.pi / period)
phi2      = -r**2
N_emp     = len(g_emp)

# Simulate the model 

def simulate_uv(theta, dgdt, burn=burn_uv):
    s, K, C_u, v0, C_v = theta
    N = len(dgdt)
    u = np.zeros(N + burn)
    v = np.zeros(N + burn)
    u[:burn] = u_emp[-1]
    v[:burn] = v_emp[-1]
    for t in range(burn + 1, N + burn):
        i      = t - burn
        neg_dg = -max(-dgdt[i-1], 0.0)                    
        du     = (1-u[t-1]) * (s - C_u*neg_dg) - K * u[t-1] * v[t-1]
        dv     = (1-v[t-1]) * (v0 + C_v*dgdt[i-1]) - K * u[t-1] * v[t-1]
        u[t]   = np.clip(u[t-1] + dt*du, 0, 1)
        v[t]   = np.clip(v[t-1] + dt*dv, 0, 1)
    return u[burn:], v[burn:]

# Sweep over C_u with fresh GDP for every (replicate, C_u)
cu_values    = np.linspace(0.0,0.5, 10)
print(cu_values)
n_replicates = 10
base_theta   = [0.0084, 2.0, None, 0.0089, 0.5]   # C_v = 2

skew_u_mat = np.zeros((len(cu_values), n_replicates))
skew_v_mat = np.zeros_like(skew_u_mat)
skew_g_mat = np.zeros_like(skew_u_mat)

for j in range(n_replicates):
    for i, C_u in enumerate(cu_values):

        # a) brand-new GDP path
        seed = j * 100 + i
        rng  = np.random.default_rng(seed)
        N    = N_emp * cycles + burn_g
        g    = np.full(N, mean_g)
        eps  = -np.abs(rng.standard_t(df_nt, size=N)) * std_dg * alpha

        for t in range(2, N):
            g[t] = (mean_g
                    + phi1 * (g[t-1] - mean_g)
                    + phi2 * (g[t-2] - mean_g)
                    + eps[t])

        g_used = np.concatenate((g[burn_g + 24:], [g[-1]]*24))
        dgdt   = np.gradient(g_used, dt)

        # b) simulate u, v
        theta        = base_theta.copy()
        theta[2]     = C_u
        u_sim, v_sim = simulate_uv(theta, dgdt)

        # last 20 full cycles to prevent transients
        window_cycles = 20
        window_len    = window_cycles * N_emp
        end   = len(u_sim)
        start = end - window_len

        u_win = u_sim[start:end] * 100            # levels (%)
        v_win = v_sim[start:end] * 100
        g_win = g_used[start:end]*100

        # we compute the skewness of the variations of the time series
        du = np.diff(u_win)
        dv = np.diff(v_win)
        dg = np.diff(g_win)

        # unbiased skew (bias=False) with zero-σ guard
        skew_u_mat[i, j] = 0.0 if du.std() < 1e-10 else skew(du, bias=False)
        skew_v_mat[i, j] = 0.0 if dv.std() < 1e-10 else skew(dv, bias=False)
        skew_g_mat[i, j] = 0.0 if dg.std() < 1e-10 else skew(dg, bias=False)

# === 4) Mean ± 2 SE ===
mean_skew_u = skew_u_mat.mean(axis=1)
se_skew_u   = skew_u_mat.std(axis=1, ddof=1) / np.sqrt(n_replicates)
mean_skew_v = skew_v_mat.mean(axis=1)
se_skew_v   = skew_v_mat.std(axis=1, ddof=1) / np.sqrt(n_replicates)
mean_skew_g = skew_g_mat.mean(axis=1)
se_skew_g   = skew_g_mat.std(axis=1, ddof=1) / np.sqrt(n_replicates)

print(mean_skew_u)


plt.figure(figsize=(8,5))
plt.errorbar(cu_values, mean_skew_u, yerr=2*se_skew_u,
             fmt='o', linestyle='none', capsize=4, label=r'$\Delta u$')
plt.errorbar(cu_values, mean_skew_v, yerr=2*se_skew_v,
             fmt='s', linestyle='none', capsize=4, label=r'$\Delta v$')
plt.errorbar(cu_values, mean_skew_g, yerr=2*se_skew_g,
             fmt='^', linestyle='none', capsize=4, label=r'$\Delta$ GDP')

plt.xlabel(r'$C_u$', fontsize=20)
plt.ylabel('Skewness', fontsize=20)
plt.xticks(fontsize=20)
plt.yticks(fontsize=20)
plt.legend(loc='best', fontsize=14)
plt.tight_layout()
#plt.savefig("/Users/antoinecyrusbecharat/Labor_Network_INET/skew_changes_vs_Cu_005_Cv0.pdf",bbox_inches="tight")
plt.show()
```
